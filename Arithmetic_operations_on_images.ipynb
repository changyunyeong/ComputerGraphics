```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/Letter - 1.png')
img2 = cv2.imread('/content/Letter - 2.png')

# Addition
img_addition = cv2.add(img1, img2)

# Subtraction
img_subtraction = cv2.subtract(img1, img2)

# Multiplication
img_multiplication = cv2.multiply(img1, img2)

# Divion
img_division = cv2.divide(img1, img2)

# Plot images
originals = {
    'Original Image 1': img1,
    'Original Image 2': img2
}

imgs = {
    'img_addition': img_addition,
    'img_subtraction': img_subtraction,
    'img_multiplication': img_multiplication,
    'img_division': img_division
}

plt.figure(figsize=(10, 4))
for i, (k, v) in enumerate(originals.items()):
    plt.subplot(1, 2, i + 1)
    plt.imshow(v[:, :, ::-1])
    plt.title(k)
    plt.xticks([]); plt.yticks([])
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))

for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(2,2, i + 1)
    plt.imshow(v[:,:,::-1])
    plt.title(k)
    plt.xticks([]); plt.yticks([])

plt.tight_layout()
plt.show()
```
